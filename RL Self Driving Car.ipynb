```python
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import os
```

# 2. Test Environment

```python
environment_name = 'CarRacing-v0'
env = gym.make(environment_name)
```

```python
env.reset()
```

```python
env.action_space
```

```python
env.observation_space
```

```python
env.render()
```

```python
env.close()
```

```python
# to understnd our environment
episodes = 5
for episode in range(1, episodes+1):
    obs = env.reset() # this gives the observations about the environment
    done = False
    score = 0
    
    while not done:
        env.render() # allows to view the environment
        action = env.action_space.sample() #generating a random action of 0 or 1 for now
        obs, reward, done, info = env.step(action) # pass through our random action to our env and we recieve info on the reward as well as a new state
        score += reward
    print('Episode:{} Score{}'.format(episode, score))
# env.close()
```

```python
env.close()
```

# 3.Train Model

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env])
```

```python
log_path = os.path.join('Training', 'Logs')
model = PPO('CnnPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=500000)
```

# 4. Save Model

```python
ppo_path = os.path.join('Training', 'Saved Models', 'PPO_2m_Driving_model')
```

```python
model.save(ppo_path)
```

```python
del model
```

```python
model = PPO.load(ppo_path, env)
```

# 5. Evaluate and Test

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
```

```python
env.close()
```

```python
# to actually test our model
episodes = 5
for episode in range(1, episodes+1):
    obs = env.reset() # gets obs of current environment
    done = False
    score = 0
    
    while not done:
        env.render() 
        action, _ = model.predict(obs) # now the action is based on the models training to predict the correct response
        obs, reward, done, info = env.step(action) # pass through our random action to our env and we recieve info on the reward as well as a new state
        score += reward
    print('Episode:{} Score{}'.format(episode, score))
# env.close()
```
